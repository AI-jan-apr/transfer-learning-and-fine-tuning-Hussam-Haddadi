```python
# Function to download and extract dataset if not already present
def get_data_extract():
    if "dataset" in os.listdir():
        print("Dataset already exists")
    else:
        print("Downloading the data...")
        !wget -O food-data.zip https://www.kaggle.com/api/v1/datasets/download/trolukovich/food11-image-dataset
        print("Dataset downloaded!")
        print("Extracting data..")
        !mkdir dataset
        !unzip -q food-data.zip -d dataset
        print("Extraction done!")

# Call the function to extract data
get_data_extract()
```

```python
# Create Dataset from list of paths
path = glob.glob("dataset/*/*.jpg")
label = [i.split(".")[0].split("/")[-2] for i in path]
```

```python
# Image augmentation for training
image_width, image_height = 224, 224  # this may need to be changed based on loaded model

aug = keras.Sequential([
    keras.layers.Resizing(image_width, image_height),
    keras.layers.Rescaling(1./255),
])
```

```python
# Dataloader function with image augmentation
def load_image(path, label, aug):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = aug(image)  # Apply augmentation
    return image, label
```

```python
# Prepare dataset using TensorFlow Dataset API

# Only proceed with dataset creation and mapping if path is not empty
if len(path) > 0:
    ds = tf.data.Dataset.from_tensor_slices((path, label))
    ds = ds.shuffle(buffer_size=len(path), reshuffle_each_iteration=False)  # disable reshuffle_each_iteration to prevent re-shuffling every iteration
    ds = ds.map(lambda p, l: load_image(p, l, aug)).batch(16).prefetch(tf.data.AUTOTUNE)
else:
    print("Warning: 'path' is empty. No images found, creating an empty dataset with string types.")
    # Create an empty dataset with the correct string dtypes to prevent TypeError in downstream operations
    ds = tf.data.Dataset.from_tensor_slices((tf.constant([], dtype=tf.string), tf.constant([], dtype=tf.string)))
    # Apply batch and prefetch even for an empty dataset to maintain consistent structure
    ds = ds.batch(16).prefetch(tf.data.AUTOTUNE)
```

## Prepare Training and Validation Splits

### Subtask:
Encode labels and split the dataset into separate training and validation sets while ensuring proper preprocessing.


**Reasoning**:
I will initialize the label mapping, encode the labels into integers, and organize the file paths and labels into training, validation, and testing splits based on the directory structure.



```python
import numpy as np

# 1. Extract unique class names and create mapping
unique_labels = sorted(list(set(label)))
label_to_index = {name: i for i, name in enumerate(unique_labels)}
index_to_label = {i: name for i, name in enumerate(unique_labels)}
print(f"Classes: {unique_labels}")

# 2. Convert label list to integer-encoded equivalent
int_labels = [label_to_index[l] for l in label]

# 3. Identify split types and 4. Create separate lists
train_paths, train_labels = [], []
val_paths, val_labels = [], []
test_paths, test_labels = [], []

for p, l in zip(path, int_labels):
    parts = p.split(os.path.sep)
    split_type = parts[1] # 'training', 'validation', or 'evaluation'

    if split_type == 'training':
        train_paths.append(p)
        train_labels.append(l)
    elif split_type == 'validation':
        val_paths.append(p)
        val_labels.append(l)
    elif split_type == 'evaluation':
        test_paths.append(p)
        test_labels.append(l)

print(f"Train samples: {len(train_paths)}")
print(f"Val samples: {len(val_paths)}")
print(f"Test samples: {len(test_paths)}")
```

**Reasoning**:
I will define the preprocessing function and create the TensorFlow datasets for training, validation, and testing as per the remaining instructions.



```python
def preprocess_image(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = aug(image)
    return image, label

# Create TensorFlow datasets
train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_ds = train_ds.shuffle(buffer_size=len(train_paths)).map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
val_ds = val_ds.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))
test_ds = test_ds.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)

# Verification
for imgs, lbls in train_ds.take(1):
    print(f'Batch image shape: {imgs.shape}')
    print(f'Batch label shape: {lbls.shape}')
    print(f'Sample labels: {lbls.numpy()[:5]}')
```

```python
import glob
import os
import tensorflow as tf

# 1. Correct the path discovery
path = glob.glob('dataset/*/*/*.jpg')
label = [p.split(os.sep)[-2] for p in path]

# 2. Encode labels
unique_labels = sorted(list(set(label)))
label_to_index = {name: i for i, name in enumerate(unique_labels)}
int_labels = [label_to_index[l] for l in label]

# 3. Organize paths and labels by split
train_paths, train_labels = [], []
val_paths, val_labels = [], []
test_paths, test_labels = [], []

for p, l in zip(path, int_labels):
    split_type = p.split(os.sep)[1]
    if split_type == 'training':
        train_paths.append(p)
        train_labels.append(l)
    elif split_type == 'validation':
        val_paths.append(p)
        val_labels.append(l)
    elif split_type == 'evaluation':
        test_paths.append(p)
        test_labels.append(l)

# 4. Create TensorFlow Datasets
def preprocess_image(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [224, 224])
    image = image / 255.0
    return image, label

train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_ds = train_ds.shuffle(len(train_paths)).map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
val_ds = val_ds.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((test_paths, test_labels))
test_ds = test_ds.map(preprocess_image).batch(32).prefetch(tf.data.AUTOTUNE)

print(f'Populated paths: {len(path)}')
print(f'Train size: {len(train_paths)}, Val size: {len(val_paths)}, Test size: {len(test_paths)}')
for imgs, lbls in train_ds.take(1):
    print(f'Successfully created train_ds. Batch shape: {imgs.shape}')
```

## Feature Extraction

### Subtask:
Implement and train an EfficientNetB0 feature extraction model with frozen weights and a custom classification head.


**Reasoning**:
I will implement the EfficientNetB0 feature extraction model as specified in the instructions, ensuring the base model is frozen, a custom classification head is added, and it is trained for 5 epochs with history saved.



```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras import layers, models

# 2. Instantiate the EfficientNetB0 base model
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# 3. Freeze the pre-trained weights
base_model.trainable = False

# 4. Create the Sequential model with custom classification head
model_fe = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.2),
    layers.Dense(11, activation='softmax')
])

# 5. Compile the model
model_fe.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 6 & 7. Train the model and save history
print("Starting Feature Extraction training for 5 epochs...")
history_fe = model_fe.fit(
    train_ds,
    epochs=5,
    validation_data=val_ds
)

print("Feature Extraction training complete.")
```

**Reasoning**:
The feature extraction experiment is complete. The next step is to perform fine-tuning by unfreezing the top layers of the EfficientNetB0 base model and continuing training with a lower learning rate to improve performance.



```python
from tensorflow.keras import optimizers

# 1. Unfreeze the base model
base_model.trainable = True

# 2. Re-compile the model with a lower learning rate for fine-tuning
model_fe.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 3. Train for 5 more epochs and save history
print("Starting Fine-tuning training for 5 epochs...")
history_ft = model_fe.fit(
    train_ds,
    epochs=5,
    validation_data=val_ds
)

print("Fine-tuning complete.")
```

## Visualize and Compare Results

### Subtask:
Generate side-by-side plots for training and validation accuracy and loss, comparing the 'Feature Extraction' (history_fe) and 'Fine-tuning' (history_ft) phases.


**Reasoning**:
I will create a visualization to compare the performance of the feature extraction and fine-tuning phases by plotting their accuracy and loss metrics side-by-side.



```python
import matplotlib.pyplot as plt

def plot_history_comparison(history_fe, history_ft):
    # Extract metrics
    acc_fe = history_fe.history['accuracy']
    val_acc_fe = history_fe.history['val_accuracy']
    loss_fe = history_fe.history['loss']
    val_loss_fe = history_fe.history['val_loss']

    acc_ft = history_ft.history['accuracy']
    val_acc_ft = history_ft.history['val_accuracy']
    loss_ft = history_ft.history['loss']
    val_loss_ft = history_ft.history['val_loss']

    epochs_fe = range(1, len(acc_fe) + 1)
    epochs_ft = range(len(acc_fe) + 1, len(acc_fe) + len(acc_ft) + 1)

    plt.figure(figsize=(14, 6))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_fe, acc_fe, color='blue', label='FE Train Acc')
    plt.plot(epochs_fe, val_acc_fe, color='blue', linestyle='--', label='FE Val Acc')
    plt.plot(epochs_ft, acc_ft, color='green', label='FT Train Acc')
    plt.plot(epochs_ft, val_acc_ft, color='green', linestyle='--', label='FT Val Acc')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_fe, loss_fe, color='red', label='FE Train Loss')
    plt.plot(epochs_fe, val_loss_fe, color='red', linestyle='--', label='FE Val Loss')
    plt.plot(epochs_ft, loss_ft, color='orange', label='FT Train Loss')
    plt.plot(epochs_ft, val_loss_ft, color='orange', linestyle='--', label='FT Val Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Execute plotting function
plot_history_comparison(history_fe, history_ft)
```
